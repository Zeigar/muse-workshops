```python
import mne
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
mne.set_log_level('WARNING')


df = pd.read_csv('reg_meditate_shrooms.csv')
```

```python
df = df.drop('AUX', 1)
df = df.drop('time', 1)
df.dropna()
```

```python
plt.plot(df['time'][:1000], df['TP9'][:1000])
```

```python
# Read the CSV file as a NumPy array
data = np.loadtxt('reg_meditate_shrooms.csv', delimiter=',', skiprows=1)
data = data.T
```

```python
# Some information about the channels
# ch_names = ['time', 'TP9', 'AF7', 'AF8', 'TP10', 'AUX']  # TODO: finish this list
ch_names = ['TP9', 'AF7', 'AF8', 'TP10']  # TODO: finish this list

# Sampling rate of the Nautilus machine
sfreq = 500  # Hz

# Create the info structure needed by MNE
info = mne.create_info(ch_names, sfreq)

# Finally, create the Raw object
raw = mne.io.RawArray(df.values.T, info)

# Plot it!
raw.plot()
```
